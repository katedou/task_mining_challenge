# This is only used for exploring data, just for showcasing one step at the really begining of model training

```python
import pandas as pd

df_parquet = pd.read_parquet("../data/uWaveGestures.parquet")
print(df_parquet)
```

```python
# check the data size/dimension and the features that is included
print("Columns of the input data : ", df_parquet.columns)
print("Dimension of the input data : ", df_parquet.shape)
```

```python
# After describe I found that there are missing values and extreme values like inf, NAN in x_Average power
df_parquet.describe()
```

# Check class distribution

```python
# check if the labels distributed euqally
labels = df_parquet.gesture.unique()
for label in labels:
    print(f"For label {label} : ")
    print(f"There are {len(df_parquet[df_parquet.gesture == label])} data points")
```

# Check missing values

```python
# Check if data contains missing values, if yes, which ones have missing values?
print("There is missing vlaue in the raw data ", df_parquet.isnull().values.any())
print("How many data points are null ", df_parquet.isnull().sum())
```

```python
# Print rows with any null values
print("Rows with null values:")
print(df_parquet[df_parquet.isnull().any(axis=1)])
```

# Check outliers

```python
import numpy as np

inf_or_neg_inf_mask = np.isinf(df_parquet) | (df_parquet == -np.inf)
print("Total number of inf or -inf values:")
print(inf_or_neg_inf_mask.sum().sum())

print("Show the top 10 rows where theere is inf or -inf in the data.")
print(df_parquet[inf_or_neg_inf_mask.any(axis=1)].head(10))
```
